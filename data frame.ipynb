# Add new column

```python
# Import data
df3 = pd.read_csv('Datasets/dog_data-short.csv')
df3
```

```python
# Via a Python List
gender = ["Male", "Male", "Female", "Male", "Female"]
df3['Gender'] = gender
df3
```

```python
######################################################## Remove column ########################################################
df3.drop('Gender', axis=1, inplace=True)
######################################################## Remove column ########################################################
```

```python
# By using DataFrame.insert()
df3.insert(4, 'Gender', gender, True) # This method forces you to specify the place of column
df3 # Does the same thing
```

```python
######################################################## Remove column ########################################################
df3.drop('Gender', axis=1, inplace=True)
######################################################## Remove column ########################################################
```

```python
# Using Dataframe.assign()
df4 = df3.assign(Gender=gender) # for preferible purposes you must use a dict instead of a list / Creates a new Dataframe
df4
```

```python
# Adding Default values
df3['Alive'] = [True]*len(df3)
df3
```

```python
# Adding a Nan Column
df3["Show Dog"] = [np.nan]*len(df3)
df3
```

# Multiple Conditional Selection

```python
df2[df2["Color"] == "BROWN"]
```

```python
len(df2[df2["Color"] == "BROWN"])
```

```python
# Multiple Conditions using AND
# In pandas we use & instead of 'and' keyword for multiple conditions
df2[(df2["Color"] == "BROWN") & (df2["Breed"] == 'MIXED')]
```

```python
# Get a count 
len(df2[(df2["Color"] == "BROWN") & (df2["Breed"] == 'MIXED')])
```

```python
# Multiple Conditions using OR
# ('or' in python) (| in pandas)
df2[(df2["Color"] == "BROWN") | (df2["Breed"] == 'MIXED')]
```

```python
# Get a count
len(df2[(df2["Color"] == "BROWN") | (df2["Breed"] == 'MIXED')])
```

```python
# Get a specific column
df2[(df2["Color"] == "BROWN") & (df2["Breed"] == 'MIXED')]["OwnerZip"]
```
